#### 1.2 Проверим, правильность расчета эффективности обогащения (вычислим её на обучающей выборке для признака rougher.output.recovery). Создадим новую колонку rougher.output.recovery_new. Для проверки найдем MAE между новой колонкой и значением признака.

Посмотрим на значения колонки rougher.output.recovery:

```python
train['rougher.output.recovery'].head(15)
```

Найдем все названия столбцов, содержание условное обозначение Aurum:

```python
for i in list_column_name_train:
    if 'au' in i:
        print(i)
```

Вспомним, что эффективность вычисляется по формуле: Recovery = ((C * (F - T)) / (F * (C - T))) * 100%

Поскольку нам нужны следующие данные:  
C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки,  

то для расчёта MAE нам нужны столбцы: *'rougher.output.concentrate_au', 'rougher.input.feed_au'* и *'rougher.output.tail_au'.*


Произведем расчеты:

```python
def rougher_output_recovery_new(row):
    C = row['rougher.output.concentrate_au']
    F = row['rougher.input.feed_au']
    T = row['rougher.output.tail_au']
    Recovery = ((C * (F - T)) / (F * (C - T))) * 100
    return Recovery
train['rougher.output.recovery_new'] = train.apply(rougher_output_recovery_new, axis=1)
train.head(10)
```

```python
rougher_output_recovery = train['rougher.output.recovery']

rougher_output_recovery_new = train['rougher.output.recovery_new']
```

```python
rougher_output_recovery.shape
```

```python
rougher_output_recovery_new.shape
```

Найдем MAE:

```python
mae = mean_absolute_error(rougher_output_recovery, rougher_output_recovery_new).round(2)
mae
```

Итак, среднее абсолютное отклонение равно нулю, соответственно расчеты, приведенные в таблице, верны. Наш новый столбец можно удалять:

```python
train = train.drop('rougher.output.recovery_new', axis=1)
```

```python
train.head()
```

На данном этапе мы подготовили данные к анализу: произвели предобработку данных, заполнив пропуски, где это возможно, и удалив их, где заполнение неточными данными повредило бы нашей работе в дальнейшем. Также мы проанализировали признаки, недоступные в тестовой выборке. Их объединяло то, что у них тип параметра "output". #### 3.1. Подготовим функцию в целях расчета окончательной sMAPE.

```python
import numpy as np
```

```python
def smape(A, B):
    smape = []
    smape = 100/len(A) * np.sum(2 * np.abs(B - A) / (np.abs(A) + np.abs(B)))  
    final_smape = 0.25 * smape[0] + 0.75 * smape[1]
    return final_smape

smape_scorer = make_scorer(smape, greater_is_better=False)
```

Теперь переходим к выбору, обучению модели и проверке ее качества посредством кросс-валидации. Будем делить выборки features_train и target_train на 5 частей.

#### Модель линейной регрессии:

Заведем необходимые параметры:

```python
param_grid_linear = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
```

Присваиваем переменную модели; используем объект GridSearchCV для поиска наилучшего набора параметров, сводящих к минимуму ошибки посредством перекрестного контроля (cross-validation):

```python
linear_regress = LinearRegression()
grid_search_linear = GridSearchCV(linear_regress,param_grid=param_grid_linear, cv=5, scoring=smape_scorer)
```

#### Модель "Дерево решений":

По результатам обучения двух моделей можно сказать, что данные схожи. Возьмем тогда модель случайного леса и посмотрим, будут ли показатели после обучения этой модели отличаться от результатов первых двух. Однако по скорости модель случайного леса явно уступит другим.

#### Модель "Случайный лес":

```python
param_grid_forest = {'max_depth': [90, 100, 110], 'n_estimators': [100, 200]}
```

```python
model = RandomForestRegressor(random_state=12345)
grid_search_forest = GridSearchCV(model, param_grid=param_grid_forest, cv=5, scoring=smape_scorer)
```

```python
grid_search_forest.fit(features_train, target_train)
```

```python
pred_forest = grid_search_forest.predict(features_train)

final_smape_forest = smape(target_train, pred_forest)

print('Final sMAPE forest: {:.4f}'.format(final_smape_forest))
```

```python
print('Best score forest = {0:.4f}; Best Parameter forest = {1}'.format(-grid_search_forest.best_score_, grid_search_forest.best_params_)) # Score is the negative because large score indicates better fit
```

Выберем Случайный лес. Не смотря на то, что эта модель обучается дольше всех, показатель sMAPE у нее на порядок ниже, чем у остальных моделей. Тем не менее score у модели также ниже (не намного, но разница присутствует). Поэтому выберем модель случайного леса в регрессии и дополнительно проверим качество модели:

```python
features_train_train, features_train_valid, target_train_train, target_train_valid = \
                                train_test_split(features_train, target_train, test_size=0.25, random_state=12345)
```

```python
final_model = RandomForestRegressor(max_depth= 90, n_estimators= 100, random_state=12345)
final_model.fit(features_train_train, target_train_train)
predicted_valid = final_model.predict(features_train_valid)
```

Теперь проверим модель на адекватность, а для этого рассчитаем коэффициент детерминации.

```python
print("R2 =", r2_score(target_train_valid, predicted_valid))
print("MAE =", mean_absolute_error(target_train_valid, predicted_valid))
```

Рассчитанный нами коэффициент детерминации можно интерпретировать как соответствие модели данным. Для приемлемых моделей предполагается, что коэффициент детерминации должен быть не меньше 50% (в этом случае коэффициент множественной корреляции превышает по модулю 70%). Наш коэффициент = 70%, что говорит о приличном качестве обучения модели. Также мы рассчитали среднее абсолютное отклонение (MAE). Оно составило 2,75. Соответственно, средняя сумма абсолютной разницы между фактическим значением и прогнозируемым значением составляет менее 3, что так же говорит в пользу выбранной модели.
